### Let's loop it out

```python
for i in range (df_new_obs.shape[0]):
    
    insert_query += '(' # with this, we are gonng start every row with row with open parenthesis
    
    for j in range(df_new_obs.shape[1]): ## This is for every column of our dataframe
        insert_query += str(df_new_obs[df_new_obs.columns.values[j]][i]) + ', ' ## We want to covert the value from each column into a String
        #
        #important: we need the values within parenthesis (xx, xx, xx) so, we do that with
        # we are skipping the last 2 characters that are the comma and the white space and changing them for "), "
        
    insert_query = insert_query[:-2] + '), '  ## And add it to our insert statement
     
```

```python
insert_query
```

### As you can see, our output doesn't end with the ' ; ' we need at the end of our SQL statement, so we are gonna fix it outside the loop

```python
insert_query = insert_query[:-2] + ';'
```

```python
insert_query
```

```python
cursor.execute(insert_query)
```
